Objective: To use Kmeans clustering to identify dominant color in an image.
Theory: Image clustering or image segmentation is the process of partitioning an image into separate segments (in our case, colors). In this project we will demonstrate how this Kmeans clustering can be used to perform color segmentation on an image. In summary we are assignming pixels that have similar colors to the same image segment. Let's get started!

### 3. load the rgb image and use reshape to convert it into a list of pixels

```python
#loading and showing image
FlagCanada = Image.open('C:\\Users\\dtosi\\canada.jpg')
plot.imshow(FlagCanada)
```

```python
FlagCanada_np = np.asarray(FlagCanada)

(h, w, c) = FlagCanada_np.shape
print('width:  ', w)
print('height: ', h)
print('channel:', c)
```

```python
#This block helps us see how the image pixels are defined upon reshaping
pixels = FlagCanada_np.reshape((FlagCanada_np.shape[0]*FlagCanada_np.shape[1], FlagCanada_np.shape[2]))
pixels
```

```python
FlagCanadaArray = FlagCanada_np.reshape((-1,3)) #This flattens it the image to (w*h, C)
FlagCanadaArray.shape   
```

### 4. Use KMeans to cluster the list of pixels into 3 clusters
       [Note: KMeans helps identify clusters in unknown data structures, where K is the number of clusters we want to identify]


```python
# Image segmentations

FlagCanada_KM = KMeans(n_clusters=3, random_state=101)
#Training the kmeans algorithm
FlagCanada_KM.fit(FlagCanadaArray)
```

### 5. Predict the cluster label of every pixel in the image and plot it back as an image.

```python
FlagCanadaLabels = FlagCanada_KM.predict(FlagCanadaArray)
FlagCanadaLabels.shape
```

```python
np.set_printoptions(threshold=sys.maxsize)
```
